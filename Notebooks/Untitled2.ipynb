```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
λ = 5.0
```

```python
n = 100
x = np.random.poisson(lam=λ, size=n)
```

```python
x
```

```python
## First estimator: sample mean
x̄ = np.mean(x)
```

```python
x̄
```

```python
## Second Estimator: left order statistic
x̃ = x[0]
```

```python
x̃
```

```python
M = 100_000
wone = np.zeros(M)
wtwo = np.zeros(M)
```

```python
for i in range(M):
    x = np.random.poisson(lam=λ, size=n)
    wone[i] = np.mean(x)
    wtwo[i] = x[0]
```

```python
wone
```

```python
wtwo
```

```python
plt.hist(wone)
```

```python
plt.show()
```

```python
plt.hist(wtwo); plt.show()
```

```python
(np.mean(wone), np.std(wone, ddof=1))
```

```python
(np.mean(wtwo), np.std(wtwo, ddof=1))
```
