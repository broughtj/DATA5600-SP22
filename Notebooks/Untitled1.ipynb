## __Poisson Example__

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
λ = 4
n = 500
data = np.random.poisson(lam=λ, size=n) # The "true" DGP
```

```python
plt.hist(data); plt.show()
```

```python
x̄ = np.mean(data) # our Estimator 
x̄
```

```python
M = 1_000_000
```

```python
xbars = np.zeros(M)
```

```python
for i in range(M): # each iteration of the loop is one sample in one dimension
    sam = np.random.poisson(lam=λ, size=n)
    xbars[i] = np.mean(sam)
```

```python
xbars
```

```python
plt.hist(xbars, bins=200);
```

```python
plt.show()
```

## __Exponential Example__

```python
λ = 4
```

```python
n = 50
```

```python
x = np.random.exponential(scale=λ, size=n)
```

$$
f(x; \frac{1}{\beta}) = \frac{1}{\beta} \exp(-\frac{x}{\beta})
$$

```python
plt.hist(x); plt.show()
```

```python
xbar = np.mean(x)
xbar
```
