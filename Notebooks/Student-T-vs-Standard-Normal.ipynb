```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

```python
n = 10_000
```

```python
z = np.random.normal(size = n)
```

```python
t = np.random.standard_t(df=n-1, size=n)
```

```python
df = pd.DataFrame({'z': z, 't': t})
```

```python
df.plot.kde()
```
